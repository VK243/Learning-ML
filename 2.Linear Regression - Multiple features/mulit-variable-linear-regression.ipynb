```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('/kaggle/input/multi-feature-linear-data/train.csv')
df
```

```python
import math
median_val = math.floor(df.bedrooms.median())
median_val
```

```python
df.bedrooms = df.bedrooms.fillna(median_val)
df
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area','bedrooms','age']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[3000,3,15]])
```
