```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df['target'] = iris.target
df
```

```python
df['flower_name'] = df.target.apply(lambda x : iris.target_names[x])
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='red',marker='+')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='blue',marker='+')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='green',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='red',marker='+')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='blue',marker='+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target','flower_name'],axis='columns')
```

```python
y = df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.target
```

```python
df = pd.DataFrame(digits['data'])
```

```python
df['target'] = digits.target
```

```python
df
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(df.drop(['target'],axis='columns'),df.target,test_size=0.2)
```

```python
mod = SVC()
mod.fit(xtrain,ytrain)
```

```python
mod.score(xtest,ytest)
```

```python
mod1 = SVC(kernel='linear')
mod1.fit(xtrain,ytrain)
```

```python
mod1.score(xtest,ytest)
```

```python
mod2 = SVC(kernel='poly')
mod2.fit(xtrain,ytrain)
```

```python
mod2.score(xtest,ytest)
```
