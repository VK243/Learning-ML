```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/kaggle/input/linear-data/train.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Area (sq. ft)")
plt.ylabel("Price (US $)")
plt.scatter(df.area,df.price,color='red',marker="+")
```

```python
linear = linear_model.LinearRegression()
linear.fit(df[['area']],df.price)
```

```python
linear.predict([[5000]])
```

```python
linear.coef_
```

```python
linear.intercept_
```

```python
pd_pre = pd.read_csv("/kaggle/input/linear-data/test.csv")

p = linear.predict(pd_pre)
```

```python
pd_pre['price'] = p
```

```python
pd_pre.to_csv("prediction.csv",index=False)
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area,df.price,color='red',marker="*")
plt.plot(df.area,linear.predict(df[['area']]),color='blue')
```
